## Math

<h3 align='center'> Equation of a Circle When the Centre is Origin</h3> 

$$ x^2 + y^2 = r^2 $$ 
$$ let\  r = 1 $$
$$ x^2 + y^2 = 1^2 $$ 

<h3 align='center'> What happens if we increase the exponents of the variables in this circle equation? </h3>

$$ x^4 + y^4 = 1^4 $$ 
$$ x^8 + y^8 = 1^8 $$ 
$$ \vdots $$ 
$$ x^\infty + y^\infty = 1^\infty $$ 

<h3 align='center'> Let's see! </h3>

## Libraries

```python
import numpy as np
from cmath import inf
import matplotlib.pyplot as plt
```

## Function & Plot

```python

exps = [2, 4, 8, inf]

def plot_graph(exp, title):
    r = 1 # radius 
    x = np.linspace(-r, r, 1000)
    y = np.sqrt(r-x**exp)

    plt.figure(figsize=(5,5), dpi=100) 
    plt.grid(linestyle='-', linewidth=2)
    plt.plot(x, y, color='g')
    plt.plot(x, -y, color='r')
    plt.axhline(y=0, color='b')
    plt.axvline(x=0, color='b')
    plt.title(title)
    plt.show()

for exp in exps:
    if type(exp) == int:
        plot_graph(exp = exp, title = f'Graph of: $x^{exp} + y^{exp} = 1^{exp}$')
    elif type(exp) == float:
        plot_graph(exp = exp, title = f'Graph of: $x^\infty + y^\infty = 1^\infty $')

```

## Conclusion

### -> When the exponents of the variables in this circle equation approach infinity, the shape of the graph changes from perfect circle to square.
